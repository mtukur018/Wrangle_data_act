## Insights:
the insight gained from the graph plotted with Mathplotlib indicates that there is a positive relationship (correlation) between the two variables Favourite count and Retweet count

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
master_data.plot(kind='scatter', x='favorite_count', y='retweet_count')
plt.title('Favorite Count V.S. Retweet Count')
plt.savefig('fav_retweet.png');
```

```python
plt.figure(figsize=(15,10))
sns.set(style="white")
sns.boxplot(x='retweet_count', y="favorite_count", data = master_data).set_title('box plot')
plt.savefig('stag_fav.png');
```
